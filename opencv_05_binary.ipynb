# 이미지 이진화(Binary)

이미지 처리에서 이진화는 어떤 주어진 임계값(threshold)보다 밝은 픽셀들은 모두 흰색으로, 

어두운 픽셀들은 모두 검은색으로 변경하는 것을 의미한다

기준값에 따라 이미지는 검정/하양 2가지 값으로만 표현된다  

```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
file = files.upload()
img = cv2.imread("flower.jpg", cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
height, width = img.shape 

binary = np.zeros(shape=(height, width), dtype=np.uint8)

for y in range(height):
     for x in range(width):
      s = img[y][x]
      thresh = 128 # 임곗값은 0과 255사이 중간인 128로 설정함

      if s > thresh:
        binary [y][x] = 255

cv2_imshow(binary)
```

# threshold()

GRAYSCALE 이미지를 Binary로 변경하는 함수

*다중 채널 이미지를 입력하면 각 채널마다 이진화 연산해서 속도가 떨어짐

retval, dst = cv2.threshold(src, thresh, maxval, type)

괄호()안은 순서대로 입력 이미지, 임곗값, 최댓값(임계값 넘었을 때 적용할 값), 임곗값 형식

설정 임곗값(retval)과 결과 이미지(dst) 값 2개 반환
>thresholding type

THRESH_BINARY 이진화

THRESH_BINARY_INV 이진화의 역변환

THRESH_TRUNC   임곗값보다 작으면 그대로, 크면  maxval로 바꿔줌

THRESH_TOZERO THRESH_TOZERO 임곗값보다 작으면 0, 크면 그대로 바꿔줌

THRESH_TOZERO_INV THRESH_TOZERO전환의 역변환

```python
# THRESH_BINARY()
retval, dst = cv2.threshold(img,128,255,cv2.THRESH_BINARY)
cv2_imshow(dst)
```

```python
# THRESH_BINARY_INV()
retval, dst = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)
cv2_imshow(dst)
```

```python
# THRESH_TRUNC()
retval, dst = cv2.threshold(img,128,255,cv2.THRESH_TRUNC)
cv2_imshow(dst)
```

```python
# THRESH_TOZERO()
retval, dst = cv2.threshold(img,128,255,cv2.THRESH_TOZERO)
cv2_imshow(dst)
```

```python
# THRESH_TOZERO_INV()
retval, dst = cv2.threshold(img,128,255,cv2.THRESH_TOZERO_INV)
cv2_imshow(dst)
```

# adaptiveThreshold()

픽셀의 주변의 값에 의해 변환하는 함수

cv2.adaptiveThreshold(src,thresh, method, flag, block_size, bias)

괄호()안은 순서대로 입력 이미지, 최댓값, 적용할 값, 평균 or 가우시안, threshold 적용 방법, 주변값의 범위에 해당하는 블록의 크기 , 보정값

```python
# adaptiveThreshold() 주변 값의 평균에 의해 변함
mean = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,3)
cv2_imshow(mean)
```

```python
#  adaptiveThreshold() 가우시안에 의해 변환함
gaus = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,3)

cv2_imshow(gaus)
```
