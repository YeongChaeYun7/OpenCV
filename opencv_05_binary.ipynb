# 이미지 이진화(Binary)

이미지 처리에서 이진화는 어떤 주어진 임계값(threshold)보다 밝은 픽셀들은 모두 흰색으로, 

어두운 픽셀들은 모두 검은색으로 변경하는 것을 의미한다

기준값에 따라 이미지는 검정/하양 2가지 값으로만 표현된다  

```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
file = files.upload()
img = cv2.imread("flower.jpg", cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
height, width = img.shape 

binary = np.zeros(shape=(height, width), dtype=np.uint8)

for y in range(height):
     for x in range(width):
      s = img[y][x]
      thresh = 128 # 임곗값은 0과 255사이 중간인 128로 설정함

      if s > thresh:
        binary [y][x] = 255

cv2_imshow(binary)
```
