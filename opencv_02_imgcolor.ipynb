OpenCV 함수를 활용하여 이미지 타입 설정하고 호출하기


```python
import cv2 
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
file = files.upload()

# BGR 컬러 타입으로 읽어오기 (기본값), shape = (rows, cols, 3)
img1 = cv2.imread("flower.jpg", cv2.IMREAD_COLOR)

# 그레이 타입으로 읽어오기, shape = (rows, cols)
img2 = cv2.imread("flower.jpg", cv2.IMREAD_GRAYSCALE)

# 파일 원본 타입 읽어오기
img3 = cv2.imread("flower.jpg", cv2.IMREAD_UNCHANGED)

cv2_imshow(img1)
cv2_imshow(img2)
cv2_imshow(img3)
```

cv2.IMREAD_COLOR 이미지 파일의 RGB 채널을 각각의 채널로 분리하여 반환

```python
import numpy as np
```

```python
height, width, channel = img1.shape
B = np.zeros(shape=(height, width, channel),dtype=np.uint8) 
G = np.zeros(shape=(height, width, channel),dtype=np.uint8) 
R = np.zeros(shape=(height, width, channel),dtype=np.uint8) 
for y in range(height):
  for x in range(width): 
    B[y][x][0] = img1[y][x][0] 
for y in range(height):
  for x in range(width):
    G[y][x][1] = img1[y][x][1] 
for y in range(height):
  for x in range(width):
    R[y][x][2] = img1[y][x][2] 

cv2_imshow(R)
cv2_imshow(G)
cv2_imshow(B)
```

cv2.IMREAD_UNCHANGED 사용 없이 RGB에서 GRAY로 변환

```python
def bgrToGray(bgr_img):
    # BGR 색상값
    b = bgr_img[:, :, 0]
    g = bgr_img[:, :, 1]
    r = bgr_img[:, :, 2]
    result = ((0.299 * r) + (0.587 * g) + (0.114 * b))
    return result.astype(np.uint8)

gray = bgrToGray(img1)
cv2_imshow(gray)
```
