```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
```

이미지 파일 불러오기

```python
file = files.upload()
img = cv2.imread("flower.jpg") # "flower.jpg"는 업로드하는 이미지 파일명. numpy 형식으로 이미지 호출
print(img.shape) # height, width, channel 크기 확인
cv2_imshow(img) # 이미지 호출
```

이미지 복사

```python
copy_img = img
cv2_imshow(copy_img)
```

```python
# numpy 활용 copy
import numpy as np

height, width, channel = img.shape
copy = np.zeros(shape=(height, width, channel),dtype=np.uint8) # uint8 = unside int 8bit = 부호 없는 8비트 정수형(255)
for y in range(height):
  for x in range(width):
    copy[y][x][0] = img[y][x][0] # Blue channel. 단일 채널 출력하면 해당 채널 색만 적용
    copy[y][x][1] = img[y][x][1] # Green channel
    copy[y][x][2] = img[y][x][2] # Red channel
cv2_imshow(copy)
```

OpenCV에서 이미지 값 0은 검정색으로 출력 

```python
img[40:50][:][:] = 0 # item에 접근하여 값을 0으로 변경. 40~50 height는 검은 줄로 출력
cv2_imshow(img)
```
