```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
file = files.upload()
img = cv2.imread("flower.jpg") # "flower.jpg"는 업로드하는 이미지 파일명. numpy 형식으로 이미지 호출
print(img.shape) # height, width, channel 크기 확인
cv2_imshow(img) # 이미지 호출
```

이미지 사이즈 줄이기

```python
height, width, channel = img.shape
s_img = np.zeros(shape=(int(height/2), int(width/2), channel),dtype=np.uint8) 
# 만들고 싶은 이미지 크기의 0으로 채워진 이미지 생성. 바탕 역할
for y in range(int(height/2)):
  for x in range(int(width/2)):
    s_img[y][x][0] = img[y*2+1][x*2+1][0] 
    s_img[y][x][1] = img[y*2+1][x*2+1][1]
    s_img[y][x][2] = img[y*2+1][x*2+1][2] 

cv2_imshow(s_img)
```

```python
# resize()
height, width, channel = img.shape
small_img = cv2.resize(img, (int(width/2), int(height/2))) # 크기 조정
cv2_imshow(small_img)
```

이미지 사이즈 늘이기

```python
# 단순히 큰 바탕에 원본값을 그대로 넣었을 경우 
height, width, channel = img.shape

b_img = np.zeros(shape=(height*2, width*2, channel), dtype=np.uint8)
# c_height, c_width, c_channel = b_img.shape

for y in range(0, height*2, 2):
    r_img = img[int(y/2)]
    r_b_img = b_img[y]
           
    for x in range(0, width*2, 2):
        b = r_img[int(x/2)][0]
        g = r_img[int(x/2)][1]
        r = r_img[int(x/2)][2]
        
        r_b_img[x][0] = b
        r_b_img[x][1] = g
        r_b_img[x][2] = r

cv2_imshow(b_img)
# 원본보다 큰 만큼 빈 공간(검정색)으로 출력 
# 빈 공간을 무슨 값으로 채울지 설정해줘야한다 -> 설정 값에 따라 이미지의 품질 결정
```

```python
# 이미지 확대시 생기는 빈 공간을 옆의 원본 이미지값을 복사하여 채운 경우
height, width, channel = img.shape
b_img = np.zeros(shape=(height*2, width*2, channel), dtype=np.uint8)

for y in range(height*2):
  for x in range(width*2):
    b_img[y][x][0] = img[int(y/2)][int(x/2)][0] 
    b_img[y][x][1] = img[int(y/2)][int(x/2)][1]
    b_img[y][x][2] = img[int(y/2)][int(x/2)][2] 

cv2_imshow(b_img)
```

```python
# resize()
height, width, channel = img.shape
big_img = cv2.resize(img, (width*2, height*2)) # 크기 조정
cv2_imshow(big_img)
```
